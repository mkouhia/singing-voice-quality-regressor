```python
import pandas as pd

segments = pd.read_csv("../SVQTD/seg_time_lists.csv")
test = pd.read_csv("../SVQTD/test.csv")
train = pd.read_csv("../SVQTD/train.csv")
dev = pd.read_csv("../SVQTD/dev.csv")

segments_clean = pd.read_parquet("../data/segments_clean.parquet")
test_clean = pd.read_parquet("../data/test_clean.parquet")
train_clean = pd.read_parquet("../data/train_clean.parquet")
dev_clean = pd.read_parquet("../data/dev_clean.parquet")
```

```python
for name, orig, cleaned in [
    ("test", test, test_clean),
    ("train", train, train_clean),
    ("dev", dev, dev_clean),
    ("segments", segments, segments_clean),
]:
    print(f"\n{name.upper()}:")
    print("-original")
    display(orig)
    print("-cleaned")
    display(cleaned)

    print("-not included")
    display(orig[~orig.index.isin(cleaned.index)])
```
